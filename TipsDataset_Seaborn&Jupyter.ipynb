# Total Bill and Tip Amount Analysis

From an obersvers persective, one would lean towards observing the total bill and the tip amount as these are undisputable. And as the example above suggests, the tip recommendations are a % of the total bill. The more money spent in the restaurant then the larger the tip, and this stands to reason as a larger party of guests with more labour intensive demands on the staff and the kitchen, the tip should be related to the total bill. In Ireland, with the service included, we don't need to concern ourselves with the issues of tipping, thankfully. 

The following few lines of code is setting up the analysis and preparing the dataset using the various tools of Python, Seaborn, Matplot lib etc. 

```python
import numpy as np #importing numpy and giving it a short name
import pandas as pd #importing pandas and giving it a short name
import matplotlib.pyplot as plt #importing matplotlib and giving it a short name
import seaborn as sns #importing seaborn and giving it a short name
sns.set(style="darkgrid") # setting the style of seaborn plots to ticks
```

The following plot, as a very basic plot isone showing the total bill along the x axis versus the amount of the tip in the y or vertical axis. Very basic and little to be garnered from it. Or is there? 

```python
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", data=tips);
```

Is there a relationship? Quite evidently there is bcause of the plot. Because the plot is not completely random, is proof there is a relationship. But to what extent. How can we further plot or analyze this data to further show to us this relationship. 

Further adding the 10, 15 and 20% value of the total bill reveals to us the majority of the scatter is between these:

```python
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", data=tips);
#plotting a 10% tip line to demonstrate any relationship
plt.plot([0, 50], [0, 5], label='10% tip', linewidth=1, color='red')
#plotting a 15% tip line to demonstrate any relationship
plt.plot([0, 50], [0, 7.5], label='15% tip', linewidth=1, color='blue')
#plotting a 20% tip line to demonstrate any relationship
plt.plot([0, 50], [0, 10], label='20% tip', linewidth=1, color='green')
plt.legend()
```

So yes, there are exceptions. But for the most part, customers are abiding by the rough 10-20% rules as outlined earlier. 

Another interesting reveal is to display the size of the tips compared to the total bill but using a plot where the size of the group eating has been recorded. 
Does it reveal anything of interest to us? On a  personal note I do not believe so. However, as a restaurant owner, they may be interested in knowing the size of groups attending their restaurant and what the total spend is per group. 

```python
sns.relplot(x="total_bill", y="tip", size="size", sizes=(3, 100), data=tips); #plot total bill vs tips with size included
```

Taking at look at only the tips versus the total bill does offer up some interesting points. See the plot below and allow your attention to be drawn to the 3 arrows added.

```python
#sns.relplot(x="total_bill", y="tip", size="size", sizes=(1, 30), data=tips); #plot total bill vs tips with size included
sns.relplot(x="total_bill", y="tip", sizes=(1, 10), data=tips);
plt.plot([22, 50], [2, 2], label='10% tip', linewidth=1, color='magenta') # put arrow in plot showing 2 dollar tips
plt.plot([22, 24], [2, 2.2], label='10% tip', linewidth=1, color='magenta') # line forming part of arrow head
plt.plot([22, 24], [2, 1.8], label='10% tip', linewidth=1, color='magenta') #

plt.plot([22, 50], [3, 3], label='10% tip', linewidth=1, color='green', linestyle='--')# put arrow in plot showing 3 dollar tips
plt.plot([22, 24], [3, 3.2], label='10% tip', linewidth=1, color='green', linestyle='--') # line forming part of arrow head
plt.plot([22, 24], [3, 2.8], label='10% tip', linewidth=1, color='green', linestyle='--') # line forming part of arrow head

plt.plot([30, 50], [4, 4], label='10% tip', linewidth=1, color='red', linestyle=':')#put arrow in plot showing 4 dollar tips
plt.plot([30, 32], [4, 4.2], label='10% tip', linewidth=1, color='red', linestyle=':') # line forming part of arrow head
plt.plot([30, 32], [4, 3.8], label='10% tip', linewidth=1, color='red', linestyle=':') # line forming part of arrow head
```

Interestingly, we can see patterns in the data. For instance, maybe for convenience, the tips are evidently more prominent, certainly in the lower value of total bill, at the whole number denominations. The magenta coloured arrow is pointing at the intense number of meals served that totalled between 10 - 20 and resulted in a 2 dollar tip. There were even some very generous patrons that paid less than 10 dollars for their meals but offered a 2 dollar tip which is flying in the face of the 10-20% recommendations earlier mentioned. 

Another interesting point to note between the total amount of the bill and the tip given is the customers decision to just round up the total including the tip to the next whole note denomination. For instance, if a meal may come to 16.93 then the amount tendered was most likely 20 (if paying by cash) and the tip in this case offered was 3.07. In this case we don't require a data analyst to check this data but certainly it is an intereting item to investigate. To check the data and look for the total amount paid including the tip, if this equals a whole number, then chances are the customer rounded it up in their heads and handed over the amount. A sense of "keep the change" as it were. 

Incidentally, one very exceptional character offered over a 5 dollar tip based on a 7-8 dollar meal. An exceptional tip making the waiting staff very happy indeed. If only more customers were this generous! 
See the following plot and the red arrow:

```python
sns.relplot(x="total_bill", y="tip", sizes=(1, 10), data=tips);
# Annotate with text + Arrow
plt.annotate(
# Label and coordinate
'Exceptional Tip!', xy=(7.6, 5.2), xytext=(1, 7),
 
# Custom arrow
arrowprops=dict(facecolor='red', shrink=0.05)
)
```

Another convenient way to analyse or to view the data is to check using a swarm plot. Conveniently seaborn has this integrated, so to do so, is relatively straight forward. A swarm plot allows us to see the distribution of values, as the data is adjusted along the categorical axis, with no overlap. Meaning if we count the dots reporesented in the plot below we will have 244. And from a brief observation we can see where the most money is being spent in the restaurant.  

```python
sns.swarmplot(x=tips["total_bill"])
```

So what does this plot tell us? Quite simply we can see that the majority of meals provided were between 20 and 30 with only a handful at the upper end between 40 and 50. 
How can the restaurant owner make use of this information? Well for one he might want to review his menu offer and remove the more expensive items. Or maybe too much emphasis has been put on the 10-20 price bracket offer and a review of the menu to expand it and the price ranges is in order. 

Sex! Yes this has been recorded as part of the data set. What possible information could we garner from knowing if the tipster was male or female? Well for one we would see if men or women are more generous. We would also see what the ratio of men to women paying for the meals is. And even further than that, taking into account the day, time, smoking habits, amount of tip, group size enad tipping habits, wellwe should be able to see quite a considerable amount. 


```python
sns.catplot(x="sex", y="tip", data=tips);
```

Well look at that! Men appear to be bigger tippers at this restaurant. However, if we plot the data to see the size of the groups where both female and males tipped, we can see that the men are tipping when there are groups attending the restaurant. Males or females appear to be similar in value of tips up to about 5 dollars. 

```python
sns.relplot(x="sex", y="tip", size="size", sizes=(3, 50), hue='sex', data=tips)
```

Further expanding the data we can include the days the patrons frequent the restaurant. Interestingly, Thursday appears to be a day that is visited more so by the female population, with all other days with a relatively good mix of both sexes. 

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=tips); #Category swarm plot - day vs total with sex in the mix
```

For box plot enthuasists, the same data can be plotted, check the plot below:

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="box", data=tips); # category box plot - total bill vs sex.
```

A very intuitive plot displaying the number of males v's females across the 4 days. Quite easily interpreted, the male customers generally paying more, especially Fridays.
